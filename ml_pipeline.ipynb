```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
num=Pipeline(steps=[("imputer",SimpleImputer(missing_values=np.nan,strategy="mean")),("scaler",StandardScaler())])
```

```python
num
```

```python
cat=Pipeline(steps=[("imputer",SimpleImputer(fill_value="missing",strategy="constant")),("onehot",OneHotEncoder(handle_unknown="ignore"))])
```

```python
cat
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer( [("catgorical",cat,
                 ["gender","city"]),("numerical",num,["age","height"])])
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe=make_pipeline(preprocessor,LogisticRegression())
```

```python
pipe
```
